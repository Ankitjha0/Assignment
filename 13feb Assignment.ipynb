exceptions are used to handle runtime errors that occur during the execution of a program. Custom exception classes should generally inherit from the built-in Exception class or one of its subclasses, and should provide clear and meaningful error messages that help in understanding the cause of the exception. Custom exception classes should also be used judiciously and only for handling exceptional cases, not as a way to control normal program floww.

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

LookupError classes such as keyerror and indexerror , are used in programing to handle specific types of error that can
occur when attempting to access elements in collection such as dictonaries ,list, arrays. these error classes provide a way to capture and handle these errors in structured manner , allowing for graceful error and handiling in the code.

```python
#keyerror
my_dict = {"name" : "Ankit", "age" : 22 , "Gender" : "Male"}
try:
    print(my_dict["city"])
except KeyError:
    print("key does not exist in the dictonary")
###Indexerror
list = [1,2,3,4,5,6]
try:
    print(list[10])
except IndexError:
    print("Index out of range")
```

Q5. Explain ImportError. What is ModuleNotFoundError?

An ImportError is a type of error that occurs in programming when there is an issue with importing a module or a package. It typically indicates that the interpreter was unable to locate and load the specified module or package during the import process.


A "ModuleNotFoundError" is a specific type of ImportError that occurs when the module or package specified in the import statement is not found by the interpreter. It is typically raised when the interpreter is unable to locate the specified module or package in any of the directories listed in the module search path.
